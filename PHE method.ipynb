```python
from fastai import *
from fastai.text import *
import string
```

```python
def _grep(x, list_to_grep):
    """takes a string (x) and checks whether any string from a given 
       list of strings (list_to_grep) exists in `x`"""
    for text in list_to_grep:
        if text.lower() in x.lower():
            return True
    return False
```

```python
df = pd.read_csv('/rds/user/trpb2/hpc-work/data/takeaway/result_2019-09-25.txt', sep='\t')
```

```python
df['BusinessName'] = df.BusinessName.str.lower()
```

```python

sandwich = ['7844']
df_sand = df.loc[df['BusinessTypeID'].isin(sandwich)].copy()
```

```python
major_only = [ '4613', '7840', '7845']
df_major_types = df.loc[df['BusinessTypeID'].isin(major_only)].copy()
major_names = ["subway", "gregg", "mcdonald", "domino", "KFC", "kentucky", "pizza hut", "burger king", "papa john"]
mask = df_major_types.BusinessName.apply(_grep, list_to_grep=major_names)
df_major =df_major_types[mask].copy()
```

```python
terms_only = [ '7846']
df_term_types = df.loc[df['BusinessTypeID'].isin(terms_only)].copy()
term_names = ["burger", "chicken", "chip", "fish bar", "pizza", "kebab", "india", "china", "chinese"]
mask = df_term_types.BusinessName.apply(_grep, list_to_grep=term_names)
df_term =df_term_types[mask].copy()
```

```python
major_or_terms = ['1', '7841']
df_major_or_terms_types = df.loc[df['BusinessTypeID'].isin(major_or_terms)].copy()
major_or_terms_names = ["burger", "chicken", "chip", "fish bar", "pizza", "kebab", "india", "china", "chinese", 
               "subway", "gregg", "mcdonald", "domino", "KFC", "kentucky", "pizza hut", "burger king", "papa john"]
mask = df_major_or_terms_types.BusinessName.apply(_grep, list_to_grep=major_or_terms_names)
df_major_or_terms =df_major_or_terms_types[mask].copy()
```

```python
df_final = pd.concat([df_sand, df_major, df_term, df_major_or_terms])
```
