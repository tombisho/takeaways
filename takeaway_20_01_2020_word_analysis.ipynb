```python
label_count = pd.DataFrame(df.groupby('label').ascii_name_fix.count().sort_values(ascending=False), )
label_count.reset_index(level=0, inplace=True)
label_count
# df2 = pd.DataFrame(np.sort(df[['label', 'label2']], axis=1), index=df.index)
# df2.columns = ['label', 'label2']
# combos = df2.groupby(['label', 'label2']).size().reset_index(name='counts')
```

```python
label1 = df.label.unique()
label2 = df.label2.unique()
labels = np.unique(np.concatenate((label1,label2), axis=0))

big_list = []
for label in labels:
    counts = Counter(_ for _ in " ".join((list(df.ascii_name_fix[(df.label==label)|(df.label2==label)]))).split()).most_common(10000)
    big_list += [[word[0],label, word[1]] for word in counts]
```

```python
big_list
```

```python
 df2 = pd.DataFrame.from_records(big_list, columns=['word', 'label', 'count'])
```

```python
df2.sort_values(by='count', ascending = False)[0:50]
```

```python
to_drop = ['the', 'and', 'of', 'restaurant', 'takeaway']
for word in to_drop:
    df2 = df2[df2['word'] != word]
```

```python
df2['comb'] = df2[['word', 'count']].values.tolist()
```

```python
#df3=pd.DataFrame(pd.pivot_table(df2, values='count', index='word', columns='label', aggfunc='count', fill_value=0).to_records())
df3=pd.DataFrame(pd.pivot_table(df2, values='count', index='word', columns='label', aggfunc='sum', fill_value=0).to_records())
```

```python
words_per_cat = pd.DataFrame(df2.groupby('label')['comb'].apply(list))
words_per_cat.reset_index(level=0, inplace=True)
words_per_cat
#df2.groupby('label').agg({ 'word': lambda x: list(x),'count': lambda x: list(x)})
# for label in label_count[:10]:
#     print(label)
```

```python
result = pd.merge(label_count,
                 words_per_cat,
                 on='label')
result
```
